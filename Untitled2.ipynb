# Exercício 3

```python
# exercício 3 da atividade prática
#------------------- início função dimensoesObjeto ------------------------
def dimensoesObjeto():
  while True: # repetição
    try: # correção de erro
      comprimentoOb = int(input('Digite o comprimento do objeto (em cm): '))
      larguraOb = int(input('Digite a largura do objeto (em cm): '))
      alturaOb = int(input('Digite a altura do objeto (em cm): '))
      volumeP = alturaOb * larguraOb * comprimentoOb # multiplicação aplicada
      print('O volume do objeto é (em cm³): {:.1f}' .format(volumeP))
      # multiplicador da dimensão
      if (volumeP < 1000):
        return 10
      elif (volumeP >= 1000) and (volumeP < 10000): 
        return 20
      elif (volumeP >= 10000) and (volumeP < 30000): 
        return 30
      elif (volumeP >= 30000) and (volumeP < 100000): 
        return 50
      elif (volumeP > 100000): 
        print('Não aceitamos objetos com dimensões tão grandes.\n'+
              'Entre com as dimensões desejadas novamente.') # alerta de incorreto
        continue # opção volta ao começo while
    except ValueError:
      print('Você digitou alguma dimensão do objeto com valor não numérico.\n'+
            'Por favor entre com as dimensões desejadas novamente.') # alerta de incorreto
      continue # opção volta ao começo while
    break # fim do laço repetir
#------------------------- Fim dimensoesObjeto ----------------------------
#-----------------------início função pesoObjeto --------------------------
def pesoObjeto():
  while True: # repetição
    try: # correção de erro
      pesoP = float(input('Digite o peso do objeto (em Kg): ')) # entrada com peso desejado
      # multiplicador do peso
      if (pesoP <= 0.1): 
        return 1
      elif (pesoP >= 0.1) and (pesoP < 1): 
        return 1.5
      elif (pesoP >= 1) and (pesoP < 10): 
        return 2
      elif (pesoP >= 10) and (pesoP < 30): 
        return 3
      elif (pesoP >= 30): 
        print('Não aceitamos objetos tão pesados.\n'+ 
            'Entre com o peso desejado novamente.') # alerta de incorreto
        continue # opção volta ao começo while
    except ValueError:
      print('Você digitou peso do objeto com valor não numérico.\n'+
            'Por favor entre com o peso desejado novamente.') # alerta de incorreto
      continue # opção volta ao começo while
    break # fim do laço repetir
#------------------------- fim pesoObjeto ---------------------------------
#-------------------- início função rotaObjeto ----------------------------
def rotaObjeto():
  while True: # repetição
      # lita das rotas
      rotaP = input('Selecione a rota: \n' 
      'SR - De São Paulo até Rio de Janeiro \n'
      'BS - De Brasília até São Paulo \n'
      'SB - De São Paulo até Brasília \n'
      'BR - De Brasília até Rio de Janeiro \n'
      'RB - Rio de Janeiro até Brasília \n'
      '>> ') 
      # multiplicador da rota
      if rotaP == 'SR': 
       return 1
      elif rotaP == 'BS': 
       return 1.2
      elif rotaP == 'SB': 
        return 1.2
      elif rotaP == 'BR':
        return 1.5
      elif rotaP == 'RB': 
        return 1.5
      else:
        print('Você digitou uma rota que não existe.\n'+
              'Por favor entre com a rota desejada novamente.') # alerta de incorreto
        continue # opção volta ao começo while
      break  # fim do laço repetir
#--------------------------- fim rotaObjeto -------------------------------
print('Bem vindo a Companhia de Logística Maike Simoncini da Silva S.A.') # identificador pessoal
dimensoes = dimensoesObjeto()
peso = pesoObjeto()
rota = rotaObjeto()
total = dimensoes * peso * rota # multiplicação aplicada
print('Total a pagar(R$): {:.2f} (dimensões: {:.2f} * peso: {:.2f} * rota: {:.2f})' .format(total,dimensoes,peso,rota)) # total

# print('------------------- Rota -----------------')
# print('---------------- Dimensões ---------------')
# print('------------------- Peso -----------------')
```

# Exercício 4

```python
# exercício 4 da atividade prática
#------------------------- início das variáveis ---------------------------
listaPeca = []
codigoPeca = 0
#-------------------------- fim das Variáveis -----------------------------
#------------------------ início cadastrar peças --------------------------
def cadastrarPeca(codigo):
  print('Você selecionou a Opção de Cadastrar Peça ') # mensagem informativa
  print('Código da Peça: {}' .format(codigo))
  nome = input('Por favor entre com o NOME da peça: ')
  fabricante = input('Por favor entre com o FABRICANTE da peça: ')
  preco = int(input('Por favor entre com o VALOR(R$) da peça: '))
  dicionarioPeca = {'codigo' : codigo, 'nome' : nome, 'fabricante' : fabricante, 'preco' : preco}
  listaPeca.append(dicionarioPeca.copy()) 
#--------------------------- fim cadastrar peças --------------------------
#------------------------- início consultar peças -------------------------
def consultarPeca():
  print('Você selecionou a Opção de Consultar Peças ') # mensagem informativa
  while True: 
    # menu secundário
    opcao_consultar = input('Escolha a opção desejada:\n'+
                            '1-Consultar Todas as Peças\n'+
                            '2-Consultar Peças por Código\n'+
                            '3-Consultar Peças por Fabricante\n'+
                            '4-Retornar\n'+
                            '>> ')
    if opcao_consultar == '1':
      print('Você Selecionou a Opção de Consultar Todas as Peças') # mensagem informativa
      for peca in listaPeca: # consultar peças na lista
        print('----------------------------------')
        for key,value in peca.items(): # fazer busca de chave e valor na lista de peças
          print('{}: {}' .format(key,value)) 
        print('----------------------------------') 
    elif opcao_consultar == '2': # consultar peças por código
      print('Você Selecionou a Opção de Consultar Peças por Código ') # mensagem informativa
      valor_desejado = int(input('Entre com o CÓDIGO desejado: '))
      for peca in listaPeca: 
        if peca['codigo'] == valor_desejado:
          print('----------------------------------')
          for key,value in peca.items():
            print('{}: {}' .format(key,value)) 
        print('----------------------------------')   
    elif opcao_consultar == '3': # consultar peças por fabricante
      print('Você Selecionou a Opção de Consultar Peças por Fabricante ') # mensagem informativa
      valor_desejado = input('Digite o FABRICANTE da Peça: ')
      for peca in listaPeca:
        if peca['fabricante'] == valor_desejado:
          print('----------------------------------')
          for key,value in peca.items():
            print('{}: {}' .format(key,value)) 
        print('----------------------------------')  
    elif opcao_consultar == '4':
      return # voltar para main
    else:
      print('Opção Inválida. Tente Novamente.') # alerta de incorreto
      continue # opção volta ao começo while
#----------------------------- fim consultar peças ------------------------
#---------------------------- início remover peças ------------------------
def removerPeca():
  print('Você Selecionou a Opção Remover peça') # mensagem informativa
  valor_desejado = int(input('Digite o código da peça a ser removida: '))
  for peca in listaPeca:
    if peca['codigo'] == valor_desejado:
      listaPeca.remove(peca)
      print('Peça Removida.') # mensagem informativa
#----------------------------- fim remover peças --------------------------
#-------------------------------- início main -----------------------------
print('Bem Vindo ao Controle de Estoque da Bicicletaria do Maike Simoncini da Silva') # identificador pessoal
while True: # repetição
  # menu principal
  opcao_principal = input('Escolha a opção desejada:\n'+ 
                          '1-Cadastrar Peças\n'+
                          '2-Consultar Peças\n'+
                          '3-Remover Peças\n'+
                          '4-Sair\n'+
                          '>> ')
  if opcao_principal == '1':
    codigoPeca = codigoPeca + 1
    cadastrarPeca(codigoPeca)
  elif opcao_principal == '2':
    consultarPeca()
  elif opcao_principal == '3':
    removerPeca()
  elif opcao_principal == '4':
    break # encerra o laço principal
  else:
    print('Opção Inválida. Tente Novamente.') # alerta de incorreto
    continue # opção volta ao começo while
#------------------------------- fim main ---------------------------------
```
